                                                        AQI Category Function

```python
def category(aqi):

    if aqi <= 50:
        return 'Good'

    elif aqi <= 100:
        return 'Satisfactory'

    elif aqi <= 200:
        return 'Moderate'

    elif aqi <= 300:
        return 'Poor'

    elif aqi <= 400:
        return 'Very Poor'

    else:
        return 'Severe'
```

                                                            Calling the Function

```python
AQI.columns = AQI.columns.str.strip()
AQI.columns
```

```python
AQI['AQI_Category'] = AQI['aqi_value'].apply(category)
AQI['AQI_Category']
```

                                                    Market Intelligence Merge

```python
city_severity = AQI.groupby('state')['aqi_value'].mean().reset_index()
popu_state = popu.groupby('state')['value'].sum().reset_index()
vehicle_state = vahan.groupby('state')['value'].sum().reset_index()
```

```python
market_data = pd.merge(city_severity, popu_state, on='state', how='inner')

market_data = pd.merge(market_data, vehicle_state, on='state', how='inner')

print(market_data.head())
```

                                                            Demand Score Formula

                                extDemandScore=0.4(extAQI)+0.3(extVehicleGrowth)+0.3(extPopulation)

```python
market_data.columns = [
    'state',
    'AQI',
    'Population',
    'Vehicles_Registered'
]

print(market_data.head())
```

```python
market_data['Demand_Score'] = (
    0.4 * market_data['AQI']
    +
    0.3 * market_data['Vehicles_Registered']
    +
    0.3 * market_data['Population']
)

print(market_data[['state', 'Demand_Score']])
```

                                                          Top Demand Cities

```python
top_demand = market_data.sort_values(
    by='Demand_Score',
    ascending=False
)

print(top_demand.head(10))
```

                                                    Demand Score Visualization

```python
print(top_demand.columns)
```

```python
plt.figure(figsize=(12,5))

sns.barplot(
    data=top_demand.head(10),
    x='state',
    y='Demand_Score'
)

plt.xticks(rotation=45)

plt.title('Top Demand States')

plt.show()
```

                                                    Per-Capita AQI Burden

                                    ext{Per-Capita AQI Burden} = rac{ ext{Average AQI}}{ ext{Population}}

```python
market_data['Per_Capita_AQI_Burden'] = (
    market_data['AQI']
    /
    market_data['Population']
)
market_data['Per_Capita_AQI_Burden'] 
```

                                                            Household Opportunity

                                            ext{Potential Households} = rac{ ext{Population}}{4}

```python
market_data['Potential_Households'] = (
    market_data['Population'] / 4
)
market_data['Potential_Households'] 
```

                                            Bubble Chart — Population vs AQI

```python
plt.figure(figsize=(10,6))

plt.scatter(
    market_data['Population'],
    market_data['AQI'],
    s=market_data['Demand_Score'] / 1000,
    alpha=0.6
)

plt.xlabel('Population')
plt.ylabel('AQI')

plt.title('Population vs AQI Demand Analysis')

plt.show()
```

                                                    Heatmap Correlation

```python
numeric_cols = market_data.select_dtypes(
    include=np.number
)

corr_matrix = numeric_cols.corr()

plt.figure(figsize=(10,6))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Heatmap')

plt.show()
```
